```python
again_q = 'y'
while again_q.lower() == 'y':
    number_grade = input(f'Please Enter a Numerical Grade: ')
    if int(number_grade) > 87:
        print('A')
    elif int(number_grade) > 79:
        print('B')
    elif int(number_grade) > 66:
        print('C')
    elif int(number_grade) > 59:
        print('D')
    else:
        print('F')

    again_q = input(f'Shall we continue (y/n)?: ')
    if again_q.lower() != 'y':
        print('Have a Great Day!')
```

```python
def remove_vowels(strng):
    return strng.replace('a','').replace('e','').replace('i','').replace('o','').replace('u','').\
        replace('A','').replace('E','').replace('I','').replace('O','').replace('U','')
```

```python
def cumulative_sum(num_list):
    running_total = 0
    cum_list = []
    
    for i in num_list:
        running_total += i
        cum_list.append(running_total)
    return cum_list
```

```python
cumulative_sum([2,3,4,5,5,5,5,5])
```
